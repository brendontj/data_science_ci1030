```
records = {}
used_labels = []
with open('AV1.txt', 'r') as reader:
  for line in reader:
    try:
      label = line.split(':')
      
      record = label[2].split('.')
      record_type = record[1]
      
      partial_label = record[2].split('-')
      family = partial_label[0]
      
      variant = partial_label[1]
    
      key = record_type + "." + family
      if key + '-' + variant not in used_labels:
        used_labels.append(key + '-' + variant)
        if key in records:
          records[key] += 1
        else:
          records[key] = 1

    except IndexError:
      continue

for k,v in sorted(records.items()):
  print(k+','+str(v))
```

```
records = []
with open('AV2.txt', 'r') as reader:
  for line in reader:
    try:
      line_splited = line.split('.')
      platform_or_language = line_splited[1]
      record_type = line_splited[2]
      rest_of_line = line_splited[3].split('-')
      sub_type = rest_of_line[0]
      variant = rest_of_line[1].split(':')[0]
      records.append(platform_or_language+','+record_type+','+sub_type+','+variant)
    except IndexError:
      continue

for i in records:
  print(i)
```
